```python
from common import data_sets, log_utils
from ch03_polynomial_regression import PolynomialRegression

num_train = 100
num_test = 100
features, poly_features, labels = data_sets.load_data_polynomial([1.2, -3.4, 5.6], 5,
                                                num_train, num_test)
```

```python
# just right
model = PolynomialRegression()
model.fit(poly_features[:num_train], labels[:num_train], lr=0.01, batch_size=10, epochs=100, 
          test_x=poly_features[num_train:], test_y=labels[num_train:])
```

```python
# under fitting
model.fit(features[:num_train], labels[:num_train], lr=0.01, batch_size=10, epochs=100,
          test_x=features[num_train:], test_y=labels[num_train:])
```

```python
# over fitting
num_samll_train = 2
model.fit(poly_features[:num_samll_train], labels[:num_samll_train], lr=0.01, batch_size=10, epochs=100, 
          test_x=poly_features[num_train:], test_y=labels[num_train:])
```
